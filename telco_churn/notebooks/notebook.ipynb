```python
import pandas as pd

df = pd.read_csv("../data/telco_churn.csv")
```

```python
df.info()

# TotalCharges is type object even though it should be float, hence there is something wrong with values

# Trying to find how many cannot be convered to numeric value
pd.to_numeric(df["TotalCharges"], errors="coerce").isna().sum()
# 11 rows have an empty string or a wrong value.
df[pd.to_numeric(df["TotalCharges"], errors="coerce").isna()]
# The value of tenure in each is 0, hence its a calculation problem.
# I know that the value will be 0 in TotalCharges, so instead of filling in median, I fill 0

df['TotalCharges'] = pd.to_numeric(df['TotalCharges'], errors="coerce").fillna(0)
# Now TotalCharges is also float64

# Checking other columns as well
df.dtypes
```

```python
# Starting Univariate and Bivariate analysis

# Understanding contract values and seeing how much there is percentage wise.
df["Contract"].value_counts(normalize=True) * 100 # type: ignore

pd.crosstab(df["Contract"], df["Churn"], normalize='index') * 100 # type: ignore

# Running on PhoneService
pd.crosstab(df["PhoneService"], df["Churn"], normalize="index") * 100

# Checking InternetService
pd.crosstab(df["InternetService"], df["Churn"], normalize="index") * 100

df["Dependents"].value_counts(normalize=True) * 100
pd.crosstab(df["Dependents"], df["Churn"], normalize="index") * 100

pd.crosstab(df["Partner"], df["Dependents"], normalize="index") * 100

df.groupby("Partner")["tenure"].mean()
```

```python
df["HasFamily"] = df["Partner"] + "_" + df["Dependents"]
print("\nHasFamily")
print(pd.crosstab(df["HasFamily"], df["Churn"], normalize="index") * 100)
print("\n\nPartner: ")
print(pd.crosstab(df["Partner"], df["Churn"], normalize="index") * 100)
print("\n\nDependents: ")
print(pd.crosstab(df["Dependents"], df["Churn"], normalize="index") * 100)

print(df["PhoneService"].unique())
print(df["InternetService"].unique())

df["ServiceInvestment"] = df["PhoneService"] + "_" + df["InternetService"]
print("\nServiceInvestment")
print(pd.crosstab(df["ServiceInvestment"], df["Churn"], normalize="index") * 100)
print("\n\nPhoneService: ")
print(pd.crosstab(df["PhoneService"], df["Churn"], normalize="index") * 100)
print("\n\nInternetService: ")
print(pd.crosstab(df["InternetService"], df["Churn"], normalize="index") * 100)


df.groupby("SeniorCitizen")["tenure"].mean()

pd.crosstab(df["SeniorCitizen"], df["Churn"], normalize="index") * 100
```

```python
# Checking more categorical columns
print(df["PaymentMethod"].value_counts(normalize=True) * 100)
print(pd.crosstab(df["PaymentMethod"], df["Churn"], normalize="index") * 100)

# Checking if automatic payments show a huge difference.
df["IsAutomaticPayment"] = df["PaymentMethod"].isin(["Bank transfer (automatic)", "Credit card (automatic)"])
print(pd.crosstab(df["IsAutomaticPayment"], df["Churn"], normalize="index") * 100)
# They don't.

# Keeping each of the 4 payment methods intact as they add more value individually.

df["TechSupport"].value_counts(normalize=True) * 100
print("\n\n")
print(pd.crosstab(df["TechSupport"], df["Churn"], normalize="index") * 100)
```
